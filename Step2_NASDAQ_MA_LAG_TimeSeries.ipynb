```python
#Helper Functions
#Source: Flatiron School

def acf_pacf(df,alags=48,plags=48):
    ''' 
        Performs acf/pacf results plot      
                             
        df          Dataframe to Analyse     
    '''

    #Create figure
    fig,(ax1,ax2) = plt.subplots(2,1,figsize=(13,8))
    #Make ACF plot
    plot_acf(df,lags=alags, zero=False,ax=ax1)
    #Make PACF plot
    plot_pacf(df,lags=plags, ax=ax2)
    plt.show()

def seasonal_plots(df,N=13,lags=[12,24,36,48,60,72]):
    ''' 
         Differencing the rolling mean to find seasonality in the resulting acf plot    
                        
         df          Dataframe to Analyse     
    '''

    fig,(ax1,ax2) = plt.subplots(2,1,figsize=(13,8))
    rolling = df - df.rolling(N).mean()
    plot_acf(rolling.dropna(),lags=lags,ax=ax1)
    plot_pacf(rolling.dropna(),lags=lags,ax=ax2)
    plt.show();
    
def train_test(df):
    ''' 
        Sets train and test data for period of prediction     
                         
        df          Dataframe to create split   
    '''
#     start_date = "2019-07-01"
#     end_date = "2020-06-30"

    #Set training data before Covid19 
    train = df[:'2020-04']
    #Set test data starting Covid19
    test = df['2020-05':]
    
    #Set training data first month Covid19 announced
    train_first_month_covid19 = df['2020-02':'2020-03']
    #Set test data
    test_first_month_covid19 = df['2020-04':'2020-05']

    #Set training data before Covid19 
    train_before_covid19 = df['2019-07':'2019-11']
    #train_before_covid19 = df[:'2019-11']
    #Set test data starting Covid19
    test_before_covid19 = df['2019-12':'2020-02']
    
    #Set training data starting into first 5 months of Covid19
    train_during_covid19 = df['2020-03':'2020-05']
    #Set test data end of Covid19 test period last month
    test_during_covid19 = df['2020-06':]
    
    #Set training data from previous year into first 5 months of Covid19
    train_before_and_during_covid19 = df[:'2020-05']
    #Set test data starting Covid19
    test_before_and_during_covid19 = df['2020-06':]
    
    
    return train, test, train_before_covid19, test_before_covid19, train_during_covid19, test_during_covid19, train_before_and_during_covid19, test_before_and_during_covid19, train_first_month_covid19, test_first_month_covid19
#     return train, test


def model_fit(df,train, test, pdq=(1,0,1),pdqs=(0,0,0,1), order=(1,2)):
    ''' 
        Performs ARIMA model given defined pdq/s values and returns results     
        
        df          Dataframe to Analyse   
        
        Returns train/test df on the same df modelled   
    '''
    

    # Fit an ARMA(1,1) model
#     model = ARIMA(ts_diff_ndx_close_ma, order=order)
    # # AR = p, I=d, MA = q, S=s
    

#     train, test = train_test(df)
    model = ARIMA(train,order=order)
    results = model.fit()
    #results.summary
    residuals = results.resid
    print(results.summary())
#     results.plot_diagnostics(figsize=(14,10))
#     plt.show();

    plt.plot(train)
    plt.plot(test)
    plt.show();
    
    
    pred = results.forecast(7)[0]
    print(pred)
       
    
    return train, test, results



def test_model(df,pdq=(1,0,1),pdqs=(0,0,0,1), display=True):
    ''' 
            Predicts for 12 months on test model with SARIMA model  
                      
            df          Dataframe to perform predictions    
    '''

    X = df.values
    train, test = X[:-12],X[-12:]
    history = [x for x in train]
    predictions = []
    for t in range(len(test)):
        model = ARIMA(train,order=pdq,seasonal_order=pdq)
        model_fit = model.fit(disp=0)
        output = model_fit.forecast()
        yhat = output[0]
        predictions.append(yhat)
        history.append(test[t])
    #rmse = sqrt(MSE(test, predictions))
    #print('SARIMA model RMSE on test data: %.5f' % rmse)
    #print('SARIMA model RMSE on test data')
    if display:
        plt.figure(figsize=(13,6))
        plt.title('Actual Test Data vs. Predictions')
        plt.plot(history[-12:],label='Actual', color='b')
        plt.plot(predictions,label='Predictions',color='r')
        plt.legend(loc='best')
        plt.show()


        
def train_model(train, results, display = True):
    ''' 
        Returns train versus test predictions pefromed on on SARIMA model    
        Transforms data into datetime object with time as index.                     

        train          Train Dataframe for period defined
        results        Results from SARIMA model
    '''

    train_pred = results.predict(-12)
#     rmse = sqrt(MSE(train[-12:],train_pred))
#     print(f'SARIMA model RMSE on train data: %.5f' % rmse)
#    print(f'SARIMA model RMSE on train data')
    if display:
        plt.figure(figsize=(13,6))
        train[-60:].plot(label='Actual',color='b')
        train_pred.plot(label='Predicted',color='r')
        plt.legend(loc='best')
        plt.title('Actual Train Data vs. Predicted Returns')
        plt.show()
        
def train_model(train, results, display = True):
    ''' 
        Returns train versus test predictions pefromed on on SARIMA model    
        Transforms data into datetime object with time as index.                     

        train          Train Dataframe for period defined
        results        Results from SARIMA model
    '''

    train_pred = results.predict(-12)
#     rmse = sqrt(MSE(train[-12:],train_pred))
#     print(f'ARIMA model RMSE on train data: %.5f' % rmse)
#    print(f'ARIMA model RMSE on train data')
    if display:
        plt.figure(figsize=(13,6))
        train[-60:].plot(label='Actual',color='b')
        train_pred.plot(label='Predicted',color='r')
        plt.legend(loc='best')
        plt.title('Actual Train Data vs. Predicted Returns')
        plt.show()
        

def forecast_model(df,order=(1,1)):
    
    ''' 
        Performs predictions on SARIMA model up to 10 years in the future   
                           
         df          Dataframe to model    
    '''

#     model = SARIMAX(df, order=pdq,seasonal_order=pdqs, enforce_stationarity=False, 
#                                         enforce_invertibility=False)
    model = ARMA(df,order=order)


    model_fit = model.fit()
    #output = model_fit.get_prediction(start='2020-05',end='2021-05', dynamic=True)
    output = model_fit.get_prediction(start=pd.to_datetime('2020-05'),end=pd.to_datetime('2021-05'), dynamic=False)
    forecast_ci = output.conf_int()
    
    if display:
        fig, ax = plt.subplots(figsize=(13,6))
        output.predicted_mean.plot(label='Forecast')
        ax.fill_between(forecast_ci.index,forecast_ci.iloc[:, 0],forecast_ci.iloc[:, 1],
                        color='k', alpha=.25,label='Conf Interval')
        plt.title('Forecast of Daily Predicted Close price by MA')
        plt.xlabel('Time')
        plt.legend(loc='best')
        plt.show()

        year_0 = output.predicted_mean['2020-05-01']
        year_1= output.predicted_mean['2021-05-01']
        year_1_ROI_100K = ((year_1 - year_0)*100)/year_0
#         year_3= output.predicted_mean['2022-05-01']
#         year_3_ROI_100K = ((year_3 - year_0)*100)/year_0
#         year_5= output.predicted_mean['2023-05-01']
#         year_5_ROI_100K = ((year_5 - year_0)*100)/year_0
#         year_10= output.predicted_mean['2024-05-01']
#         year_10_ROI_100K = ((year_10 - year_0)*100)/year_0
        
        print(f'Predicted property mean value at initial investment date 2020-05: {round(year_0,2)}')
        #print(f'Total expected return in 1 year: {round(year_1*100,2)}%')
        print(f'Predicted property mean value in 1 year: {round(year_1,2)} is {round(year_1_ROI_100K,2)} %ROI')
#         print(f'Predicted property mean value in 3 years: {round(year_3,2)} is {round(year_3_ROI_100K,2)} %ROI')
#         print(f'Predicted property mean value in 5 years: {round(year_5,2)} is {round(year_5_ROI_100K,2)} %ROI')
#         print(f'Predicted property mean value in 10 years: {round(year_10,2)} is {round(year_10_ROI_100K,2)} %ROI')

#         tot_ret = [zc,year_1,year_3,year_5,year_10]
    tot_ret = [zc,year_1,year_3,year_5,year_10]
    return tot_ret


def AIC_PDQS(df):   
    ''' 
         Runs grid search to return lowest AIC result for permutataions of pdq/s values in range 0,2  
                          
         df          Dataframe to anlyse for best pdq/s permutation    
    '''
    
    # Define the p, d and q parameters to take any value between 0 and 2
    p = d = q = range(0, 2)
    
#     Auto-Regressive (p) -> Number of autoregressive terms.
#     Integrated (d) -> Number of nonseasonal differences needed for stationarity.
#     Moving Average (q) -> Number of lagged forecast errors in the prediction equation.

    # Generate all different combinations of p, q and q triplets
    pdq = list(itertools.product(p, d, q))

    # Generate all different combinations of seasonal p, q and q triplets
    pdqs = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]


    # Run a grid with pdq and seasonal pdq parameters calculated above and get the best AIC value
    ans = []
    for comb in pdq:
        for combs in pdqs:
            try:
                mod = sm.tsa.statespace.SARIMAX(df,
                                            order=comb,
                                            seasonal_order=combs,
                                            enforce_stationarity=False,
                                            enforce_invertibility=False)

                output = mod.fit()
                ans.append([comb, combs, output.aic])
                print('ARIMA {} x {}12 : AIC Calculated ={}'.format(comb, combs, output.aic))
            except:
                continue
            
    # Find the parameters with minimal AIC value
    ans_df = pd.DataFrame(ans, columns=['pdq', 'pdqs', 'aic'])
    print(ans_df.loc[ans_df['aic'].idxmin()])
    return ans_df

def validate_stationarity(df):
    '''           
         Returns D.Fuller stationarity test results    
         
         df          Dataframe to test for stationarity    
    '''

    results = adfuller(df.dropna())
    print(f'ADFuller test p-value for ma: 2')
    print('p-value:',results[1])
    print('p-value:',results)
    if results[1]>0.05:
        print('Fail to reject the null hypothesis. Data is not stationary.\n')
    else:
        print('Reject the null hypothesis. Data is stationary.\n')


def make_datetime_index(df,col_to_make_index = 'Date', drop=False, verbose=True):
    '''           
         Transforms data into datetime object with time as index, defaulted as monthly     
         
         df          Dataframe to transform     
    '''

    df[col_to_make_index] = pd.to_datetime(df[col_to_make_index], errors='coerce')
    df = df.set_index('Date', drop=False)
    if verbose:
        display(df.index)
    return df
```

The model returns a lot of information, but we'll focus only on the table of coefficients. The coef column above shows the importance of each feature and how each one impacts the time series patterns. The $P>|z|$ provides the significance of each feature weight.

For our time-series, we see that each weight has a p-value lower or close to 0.05, so it is reasonable to retain all of them in our model.

Next, we shall run model diagnostics to ensure that none of the assumptions made by the model have been violated.

The purpose here is to ensure that residuals remain uncorrelated, normally distributed having zero mean. In the absence of these assumptions, we can not move forward and need further tweaking of the model.

Let's check for these assumptions from diagnostics plots.

In the top right plot, we see that the red KDE line follows closely with the N(0,1) line (where N(0,1)) is the standard notation for a normal distribution with mean 0 and standard deviation of 1). This is a good indication that the residuals are normally distributed.

The qq-plot on the bottom left shows that the ordered distribution of residuals (blue dots) follows the linear trend of the samples taken from a standard normal distribution with N(0, 1). Again, this is a strong indication that the residuals are normally distributed.

The residuals over time (top left plot) don't display any obvious seasonality and appear to be white noise. This is confirmed by the autocorrelation (i.e. correlogram) plot on the bottom right, which shows that the time series residuals have low correlation with lagged versions of itself.

These observations lead us to conclude that our model has no correlations and provides a satisfactory fit to help forecast future values.
